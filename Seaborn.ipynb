Seaborn is a Python data visualization library built on top of Matplotlib. It makes it easier to create statistical graphics with beautiful default styles and color palettes. Essentially, it’s used to visualize data patterns, distributions, and relationships in a more attractive and concise way than plain Matplotlib.

Use Pandas → for handling and analyzing structured data, e.g., managing patient records with blood pressure, heart rate, and age.

Use NumPy → for fast numerical computations, e.g., calculating mean and standard deviation of patients’ blood sugar levels.

Use Seaborn → for quick, beautiful statistical/EDA plots with minimal code, e.g., plotting a patient’s blood pressure distribution across age groups.

Use Matplotlib → for fully customized, low-level, or non-statistical plots, e.g., plotting a detailed heart rate trend over time for an individual patient.

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names() # all the datasets in seaborn
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins["species"].value_counts()
penguins["island"].value_counts()
```

  **Using set style **

examples:  white, dark, whitegrid,darkgrid, ticks
***Using set context***
overwrite default parameters  cahnge size of the labels, elements, text and lines
Examples:
1. Paper: Small elements ideal for prininting in academic paper
2. Notebook: Ideal for all the graphs good in google collab
3. Talk: good for presentations
4. Poster: for large display and posters


Palette in set context
1. Set1 bright and high contrast colors
2. set2 softer and pastel
3. set3 a mix of soft and bold colors

* more parameters like: deep, muted, bright, pastel, dark, colorblind

Palette according to the graph
1. Qualitataive
2. Sequential: low to high vice versa order
3. Diverging: negative, positives changes in the graph i.e profit and loss or temp changes

```python
sns.set_style("ticks")
sns.set_context("notebook")
sns.scatterplot(data = penguins, x="flipper_length_mm", y="body_mass_g", hue = "island", palette= "Set1") # used to separate data by categories with different colors in a plot and to add a third category
sns.despine(left = True) # the axis lines on the graphs are removed
```

```python
# scatterplot
sns.scatterplot(data = penguins, x ="species", y = "body_mass_g", hue ="island", style ="sex", alpha = 0.5) # alpha adjust the dataplots
```

```python
# stripplot
sns.stripplot(data = penguins, x ="species", y = "body_mass_g", hue ="island", dodge = True, jitter = True) # to speard the data points
```

```python
# Swarm plot
sns.set_context("talk")
sns.swarmplot(data = penguins, x ="species", y= "body_mass_g", hue ="island")
sns.despine()
```

```python
# histogram
sns.set_context("notebook")
sns.histplot(data = penguins, x ="body_mass_g", hue = "sex")
```
